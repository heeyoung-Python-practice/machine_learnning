```python
import pandas as pd
import numpy as np

movies = pd.read_csv("./data/movies.csv")
ratings = pd.read_csv("./data/ratings.csv")
```

```python
movies
```

```python
ratings
```

```python
# movieId 기준으로 평점 데잍터와 영화 정보 데이터를 병합
rating_movie = pd.merge(ratings, movies, on='movieId') 

rating_movie
```

```python
# 피벗데이터 색상 ( 인덱스 : 아이템, 컬럼 : 유저)
rating_matrix = rating_movie.pivot_table(
    index='title',      # 행 : 영화 제목
    columns='userId',   # 열 : 사용자 ID
    values='rating'     # 값 : 평점
).fillna(0)

rating_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity  # consime 유사도 계산 도구

item_sim = cosine_similarity(
    rating_matrix,
    rating_matrix
)   # 영화간에 유사도 게산 (아이템 -아이템끼리)

item_sim
```

```python
rating_matrix.index
```

```python
item_sim_df = pd.DataFrame(
    item_sim,                       # 영화끼리의 유사도 계산 배열
    index=rating_matrix.index,      # 행 : 영화 제목
    columns=rating_matrix.index     # 열 : 여화 제목
)

item_sim_df     # 영화 끼리의 유사도 데이터프레임
```

```python
item_sim_df.index
```

```python
item_sim_df['Toy Story (1995)'].sort_values(ascending=False)[1:11]    # Toy Story와 가장 비슷한 영화 10개
```

```python
item_sim_df['Toy Story 2 (1999)'].sort_values(ascending=False)[1:11]    # Toy Story와 가장 비슷한 영화 10개
```

```python
# 영화 추천해주는 함수
def movie_recommendations(movie_name,n):
    movie_sim = item_sim_df[movie_name].sort_values(ascending=False)[1:n+1]
    return movie_sim    # 영화 제목만 반환
```
