```python
from surprise import (SVD,          # 행렬 분해에 사용할 모델
                      Dataset,      # surprise에서 데이터에 생성
                      Reader,       # 평점 데이터 형식 정의
                      accuracy)     # 평가 지표
from surprise.model_selection import train_test_split
from surprise.dataset import DatasetAutoFolds           # 외부파일 로드 용
import pandas as pd
```

```python
df = pd.read_csv('./data/ratings.csv')
df
```

```python
df.to_csv("ratings_nohead.csv", index=False, header=False)
```

```python
df['rating'].unique()  # 평점의 고유값
```

```python
reader = Reader(
    line_format="user item rating",   #실제로 들어가는 컬럼
    sep=',',                              # 구분자(csv파일의 구분자는 .)
    rating_scale=(0.5,5)             # 범위
)

# 원본파일, raader 객체 첨부
data_fords = DatasetAutoFolds(ratings_file= 'ratings_nohead.csv', reader=reader)
```

SVD모델은 사용자-아이템-평점 행렬을 가지고 와서 잘게 분해해서 그 안에 숨겨진 취향을 찾아서 추천해주는 모델

실제 데이터를 보면
|사용자|영화A|영화B|영화C|...|
|-----|-----|-----|-----|-----|
|유저1| 4.0 | null | 3.0| ...|
|...| ... | ... | ... | ... |
|유저100| 3.5 | 4.0 | null| ...|

사용자 - 아이템 행렬
=> 사용자 잠재요인 X 아이템 잠재요인

사용자별로 액션 좋아함 or 액션 성향 pr 감성 좋아함 or 로맨스 싫어함 or 평점 주는 것을 좋아함

```python
# 데이터 쪼개기
trainset = data_fords.build_full_trainset()

model = SVD(
    n_factors = 50,
    random_state=0
)

model.fit(trainset) # 모델 학습

trainset
```

```python
model.predict('1','1')     # 사용자 '1'이 영화 '1'에게 줄 평점 예측
```
